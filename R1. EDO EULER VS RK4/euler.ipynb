```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# INITIAL CONDITIONS
x=10
v=0

# PARAMETERS
k=1
dt=0.1
m=5
t=500
```

```python
tl=[]

xle=[]
vle=[]
Ecle=[]
Eple=[]
Etle=[]

xlRK4=[]
vlRK4=[]
EclRK4=[]
EplRK4=[]
EtlRK4=[]

xe=x
xRK4=x

ve=v
vRK4=v

A=int(t/dt)

for h in range(A):
    t=h*dt
    tl.append(t)
    
    xle.append(xe)
    vle.append(ve)

    xlRK4.append(xRK4)
    vlRK4.append(vRK4)

    k1=dt*vRK4
    l1=-dt*k/m*xRK4

    k2=dt*(vRK4+l1/2)
    l2=-dt*k/m*(xRK4+k1/2)

    k3=dt*(vRK4+l2/2)
    l3=-dt*k/m*(xRK4+k2/2)

    k4=dt*(vRK4+l3)
    l4=-dt*k/m*(xRK4+k3)

    Ece=m*ve**2/2
    Epe=k*xe**2/2
    Ete=Ece+Epe

    EcRK4=m*vRK4**2/2
    EpRK4=k*xRK4**2/2
    EtRK4=EcRK4+EpRK4
    
    Ecle.append(Ece)
    Eple.append(Epe)
    Etle.append(Ete)
    EclRK4.append(EcRK4)
    EplRK4.append(EpRK4)
    EtlRK4.append(EtRK4)
    

    xe_new = xe + ve*dt
    ve_new = ve - k*xe*dt/m

    xe = xe_new
    ve = ve_new

    xRK4=xRK4 + (k1+2*k2+2*k3+k4)/6
    vRK4=vRK4 + (l1+2*l2+2*l3+l4)/6

    



    
```

```python
fig, axs = plt.subplots(1, 2, figsize=(25,8))

axs[1].plot(tl, xlRK4, label="x(t)")
axs[1].plot(tl, vlRK4, label="v(t)")
axs[1].plot(tl, EclRK4, label="Ec(t)")
axs[1].plot(tl, EplRK4, label="Ep(t)")
axs[1].plot(tl, EtlRK4, label="Et(t)")
axs[1].axhline(0, linestyle='--')
axs[1].set_title("RK4 Harmonic Oscillator")
axs[1].set_ylabel("Value")
axs[1].set_xlabel("Time")
axs[1].legend()

axs[0].plot(tl, xle, label="x(t)")
axs[0].plot(tl, vle, label="v(t)")
axs[0].plot(tl, Ecle, label="Ec(t)")
axs[0].plot(tl, Eple, label="Ep(t)")
axs[0].plot(tl, Etle, label="Et(t)")
axs[0].axhline(0, linestyle='--')
axs[0].set_title("Euler Harmonic Oscillator")
axs[0].set_xlabel("Time")
axs[0].set_ylabel("Value")
axs[0].legend()


plt.show()
```
