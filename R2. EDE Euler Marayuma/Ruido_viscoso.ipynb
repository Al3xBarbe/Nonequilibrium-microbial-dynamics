```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# INITIAL CONDITIONS
x=10
v=0

# PARAMETERS
k=1
dt=0.001
m=5
t=1000
sigma=1
gamma=0.1
NS=10
```

```python
tl=[]
xle=[]
vle=[]
Ecle=[]
Eple=[]
Etle=[]

xe=x
ve=v

A=int(t/dt)

for h in range(A):
    dW=np.random.normal(0,1)*np.sqrt(dt)
    t=h*dt
    tl.append(t)
    
    xle.append(xe)
    vle.append(ve)

    Ece=m*ve**2/2
    Epe=k*xe**2/2
    Ete=Ece+Epe
    
    Ecle.append(Ece)
    Eple.append(Epe)
    Etle.append(Ete)

    xe_new = xe + ve*dt
    ve_new = ve - k*xe*dt/m + sigma/m*dW - gamma/m*ve*dt


    xe = xe_new
    ve = ve_new

Etm=np.mean(Etle)
vem=np.mean(vle)
xem=np.mean(xle)
xvar=np.var(xle)
vvar=np.var(vle)
        
```

```python
fig, axs = plt.subplots(1, 2, figsize=(40,8))

axs[0].plot(tl, xle, label="x(t)")
axs[0].plot(tl, vle, label="v(t)")
axs[0].axhline(0, linestyle='--')
axs[0].set_title("Euler-Matuyama Harmonic Oscillator")
axs[0].set_xlabel("Time")
axs[0].set_ylabel("Value")
axs[0].legend()

axs[1].plot(tl, Etle, label="Et(t)")
axs[1].axhline(Etm, linestyle='--', label='Et_m')
axs[1].set_title("Euler-Maruyama Harmonic Oscillator")
axs[1].set_xlabel("Time")
axs[1].set_ylabel("Value")
axs[1].legend()


plt.show()
print('mean_x =',xem, '\t', 'mean v_=',vem)
print('var_x =',xvar, '\t', 'var v_=',vvar)
```
