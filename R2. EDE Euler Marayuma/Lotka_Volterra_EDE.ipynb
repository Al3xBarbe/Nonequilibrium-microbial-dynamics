```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# INITIAL CONDITIONS
x=10
v=0

# PARAMETERS
k=1
dt=0.001
m=5
t=400
sigma=10
gamma=0.01
NS=10
```

```python
tl=[]
xle=[]
vle=[]
Ecle=[]
Eple=[]
Etle=[]

xe=x
ve=v

A=int(t/dt)

for h in range(A):
    dW=np.random.normal(0,1)*np.sqrt(dt)
    t=h*dt
    tl.append(t)
    
    xle.append(xe)
    vle.append(ve)

    Ece=m*ve**2/2
    Epe=k*xe**2/2
    Ete=Ece+Epe
    
    Ecle.append(Ece)
    Eple.append(Epe)
    Etle.append(Ete)

    xe_new = xe + ve*dt
    ve_new = ve - k*xe*dt/m + sigma/m*dW - gamma/m*ve*dt


    xe = xe_new
    ve = ve_new

Etm=np.mean(Etle)
vem=np.mean(vle)
xem=np.mean(xle)
xvar=np.var(xle)
vvar=np.var(vle)
        
```

```python
fig, axs = plt.subplots(1, 1, figsize=(25,8))

axs.plot(tl, xle, label="x(t)")
axs.plot(tl, vle, label="v(t)")
axs.axhline(0, linestyle='--')
axs.set_title("Euler Harmonic Oscillator")
axs.set_xlabel("Time")
axs.set_ylabel("Value")
axs.legend()


plt.show()
print('mean_x =',xem, '\t', 'mean v_=',vem)
print('var_x =',xvar, '\t', 'var v_=',vvar)
```
