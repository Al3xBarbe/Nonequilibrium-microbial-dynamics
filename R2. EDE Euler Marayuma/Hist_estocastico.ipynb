```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# INITIAL CONDITIONS
x=10
v=0

# PARAMETERS
k=1
dt=0.001
m=5
t=1000
sigma=1
gamma=0.1
NS=10
```

```python
def hist(t,dt,x0,v0,M,sigma):
    A=int(t/dt)
    xl = []
    vl = []
    for i in range(M):
        x=x0
        v=v0
        for h in range(A):
            dW = np.random.normal(0,1)*np.sqrt(dt)
            x_new = x + v*dt
            v_new = v - k*x*dt/m - gamma/m*v*dt + sigma/m*dW

            x = x_new
            v = v_new
        xl.append(x)
        vl.append(v)
        
    plt.figure()

    plt.hist(xl, bins=100, density=True, alpha=0.6)
    plt.title("Histograma de posiciones x")

    plt.figure()

    plt.hist(vl, bins=100, density=True, alpha=0.6)
    plt.title("Histograma de velocidades v")

    plt.show()
    
            
```

```python
hist(100,0.5,10,0,1000)
```
